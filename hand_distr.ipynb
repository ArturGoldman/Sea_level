```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#open_file = "output.txt"
open_file = "output_every.txt"
```

```python
def draw_dist(a, amnt, breaker, name):
    a.sort()
    name = str(name)
    mx_val = a[-1]
    if abs(a[0]) > a[-1]:
        mx_val = abs(a[0])
    plt.figure(figsize=(12,8))
    #x, y = build_distr(raw_data10, raw_data10[0], raw_data10[-1], 200)
    (y, bins, patches) = plt.hist(a, bins=amnt, label='hst', density=True, range = (-mx_val, mx_val))
    x = list()
    #x.append((a[0] + bins[0])/2)
    for i in range(1, len(bins)):
        x.append((bins[i] + bins[i-1])/2)
    #x.append((bins[-1]+a[-1])/2)
    plt.figure(figsize=(12,8))
    plt.title("Value counts, "+ name +" m/s", fontsize=18)
    plt.yscale('log')
    plt.plot(x, y)
    plt.xticks(fontsize=18)
    plt.yticks(fontsize=18)
    
    plt.xlabel('avg_storm - avg_calm', fontsize=18)
    plt.ylabel('log normalized number of occurrences', fontsize=18)

    plt.grid()
    
    
    plt.figure(figsize=(12,8))
    left_x = list()
    left_y = list()
    right_x = list()
    right_y = list()
    for i in range(len(x)):
        if x[i] < 0:
            left_x.append(-x[i])
            left_y.append(y[i])
        else:
            right_x.append(x[i])
            right_y.append(y[i])
    plt.plot(left_x, left_y, '-b', label = "Negative values")
    plt.plot(right_x, right_y, '-r', label = "Positive values")
    plt.xlabel('avg_storm - avg_calm', fontsize=18)
    plt.ylabel('log normalized number of occurrences', fontsize=18)

    plt.yscale('log')
    plt.xticks(fontsize=18)
    plt.yticks(fontsize=18)
    plt.title("Symmetrical view of Value counts, "+ name +" m/s", fontsize=18)
    plt.grid()
    plt.legend(prop={'size': 20})
    
    plt.figure(figsize=(12,8))
    left_x = list()
    left_y = list()
    right_x = list()
    right_y = list()
    for i in range(len(x)):
        if abs(x[i]) < breaker:
            if x[i] < 0:
                left_x.append(-x[i])
                left_y.append(y[i])
            else:
                right_x.append(x[i])
                right_y.append(y[i])
    plt.plot(left_x, left_y, '-bo', label = "Negative values")
    plt.plot(right_x, right_y, '-ro', label = "Positive values")
    plt.xlabel('avg_storm - avg_calm', fontsize=18)
    plt.ylabel('log normalized number of occurrences', fontsize=18)

    plt.yscale('log')
    plt.xticks(fontsize=18)
    plt.yticks(fontsize=18)
    plt.title("Closer Symmetrical view of Value counts, "+ name +" m/s", fontsize=18)
    plt.grid()
    plt.legend(prop={'size': 20})
    
    
    plt.figure(figsize=(12,8))
    left_x = list()
    left_y = list()
    right_x = list()
    right_y = list()
    for i in range(len(x)):
        if abs(x[i]) > breaker:
            if x[i] < 0:
                left_x.append(-x[i])
                left_y.append(y[i])
            else:
                right_x.append(x[i])
                right_y.append(y[i])
    plt.plot(left_x, left_y, color = 'b', label = "Negative values")
    plt.plot(right_x, right_y, color = 'r', label = "Positive values")
    plt.xlabel('avg_storm - avg_calm', fontsize=18)
    plt.ylabel('log normalized number of occurrences', fontsize=18)

    plt.yscale('log')
    plt.xticks(fontsize=18)
    plt.yticks(fontsize=18)
    plt.title("Closer Symmetrical view of Probability Density Function, "+ name +" m/s", fontsize=18)
    plt.grid()
    plt.legend(prop={'size': 20})
```

```python
def log_dist(a, starter, multiplier, breaker, name):
    a.sort()
    name = str(name)
    mx_val = a[-1]
    if abs(a[0]) > a[-1]:
        mx_val = abs(a[0])
        
        
    my_ticks = list()
    my_ticks.append(0)
    curpos = 0
    curlen = starter
    while curpos < mx_val:
        curpos += curlen
        my_ticks.append(curpos)
        my_ticks.append(-curpos)
        curlen *= multiplier
    
    my_ticks.sort()
    
    plt.figure(figsize=(12,8))
    #x, y = build_distr(raw_data10, raw_data10[0], raw_data10[-1], 200)
    (y, bins, patches) = plt.hist(a, bins=my_ticks, label='hst', density=True, range = (-mx_val, mx_val))
    x = list()
    #x.append((a[0] + bins[0])/2)
    for i in range(1, len(bins)):
        x.append((bins[i] + bins[i-1])/2)
    #x.append((bins[-1]+a[-1])/2)
    plt.figure(figsize=(12,8))
    plt.title("Value counts, "+ name +" m/s", fontsize=18)
    plt.yscale('log')
    plt.plot(x, y)
    plt.xticks(fontsize=18)
    plt.yticks(fontsize=18)
    
    
    
    plt.xlabel('starting = '+str(starter)+', multiplier = '+str(multiplier), fontsize=18)
    plt.ylabel('log normalized number of occurrences', fontsize=18)

    plt.grid()
    
    
    plt.figure(figsize=(12,8))
    left_x = list()
    left_y = list()
    right_x = list()
    right_y = list()
    for i in range(len(x)):
        if x[i] < 0:
            left_x.append(-x[i])
            left_y.append(y[i])
        else:
            right_x.append(x[i])
            right_y.append(y[i])
    plt.plot(left_x, left_y, '-b', label = "Negative values")
    plt.plot(right_x, right_y, '-r', label = "Positive values")
    plt.xlabel('log view, starting = '+str(starter)+', multiplier = '+str(multiplier), fontsize=18)
    plt.ylabel('log normalized number of occurrences', fontsize=18)

    plt.yscale('log')
    plt.xscale('log')
    plt.xticks(fontsize=18)
    plt.yticks(fontsize=18)
    plt.title("Symmetrical view of Value counts, "+ name +" m/s", fontsize=18)
    plt.grid()
    plt.legend(prop={'size': 20})
    
    plt.figure(figsize=(12,8))
    left_x = list()
    left_y = list()
    right_x = list()
    right_y = list()
    for i in range(len(x)):
        if abs(x[i]) < breaker:
            if x[i] < 0:
                left_x.append(-x[i])
                left_y.append(y[i])
            else:
                right_x.append(x[i])
                right_y.append(y[i])
    plt.plot(left_x, left_y, '-bo', label = "Negative values")
    plt.plot(right_x, right_y, '-ro', label = "Positive values")
    plt.xlabel('log view starting = '+str(starter)+', multiplier = '+str(multiplier), fontsize=18)
    plt.ylabel('log normalized number of occurrences', fontsize=18)

    plt.yscale('log')
    plt.xscale('log')
    plt.xticks(fontsize=18)
    plt.yticks(fontsize=18)
    plt.title("Closer Symmetrical view of Value counts, "+ name +" m/s", fontsize=18)
    plt.grid()
    plt.legend(prop={'size': 20})
    
    
    plt.figure(figsize=(12,8))
    left_x = list()
    left_y = list()
    right_x = list()
    right_y = list()
    for i in range(len(x)):
        if abs(x[i]) > breaker:
            if x[i] < 0:
                left_x.append(-x[i])
                left_y.append(y[i])
            else:
                right_x.append(x[i])
                right_y.append(y[i])
    plt.plot(left_x, left_y, color = 'b', label = "Negative values")
    plt.plot(right_x, right_y, color = 'r', label = "Positive values")
    plt.xlabel('starting = '+str(starter)+', multiplier = '+str(multiplier), fontsize=18)
    plt.ylabel('log normalized number of occurrences', fontsize=18)

    plt.yscale('log')
    plt.xticks(fontsize=18)
    plt.yticks(fontsize=18)
    plt.title("Closer Symmetrical view of Probability Density Function, "+ name +" m/s", fontsize=18)
    plt.grid()
    plt.legend(prop={'size': 20})
```

```python
#lst = list()
raw_data10 = list()
raw_data15 = list()
raw_data20 = list()
f = open(open_file, "r")
for line in f:
    arr = line.rstrip().split("\t")
    if int(arr[0][:4]) >= 2011 or int(arr[0][:4]) < 2011:
        if int(arr[1]) == 10:
            raw_data10.append(float(arr[2]))
        if int(arr[1]) == 15:
            raw_data15.append(float(arr[2]))
        if int(arr[1]) == 20:
            raw_data20.append(float(arr[2]))
f.close()
```
